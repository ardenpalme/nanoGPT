```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("results.csv")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
sns.lineplot(df, x="dp", y="perplexity", ax=ax, marker='s')
ax.set_xlabel("head size")
ax.set_title("Low-rank Bottleneck")
plt.savefig("images/lrb_plot.png")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
img = mpimg.imread('images/lrb.png')
ax.imshow(img)
ax.axis('off')  # hide axes around image
ax.set_title('Transformer Performance')
plt.savefig("images/lrb_paper.png")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot([4, 8, 16, 32], [8.735513687133789, 8.424489974975586, 8.175213813781738, 7.766679763793945])
ax[0].set_xlabel('head size')
ax[0].set_ylabel('perplexity')
ax[0].set_title('Performance vs Head Size')

img = mpimg.imread('images/lrb.png')
ax[1].imshow(img)
ax[1].axis('off')  # hide axes around image
ax[1].set_title('Paper Figure 1')

plt.tight_layout()
plt.savefig('comparison.png')
```
